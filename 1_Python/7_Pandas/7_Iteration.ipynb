```python
import pandas as pd
```

### Row Wise

```python
data = pd.DataFrame({
    'name' : ["Sumit","Shivam","Saket"],
    'age' : [21,13,16],
    'city' : ["Bangaluru" , "Pune" , "Hydrabad"]
})

for index , row in data.iterrows():
    print("Row: ",index)
    print(row['name'] , row['age'] , row['city'] ,'\n')
    print(row,'\n\n')
```

```python
for row in data.itertuples():
    print(row)
    print("Index:",row.Index)
    print(row.name)
    print(row.age)
    print(row.city)
    print()
```

### Column Wise

```python
for col in data.columns:
    print("Columns:",col)
```

```python
for col in data.columns:
    print(data[col],'\n')
```

```python
for col in data.columns:
    print(data[col].values,'\n')
```
