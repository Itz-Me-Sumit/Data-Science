```python
import pandas as pd

df = pd.DataFrame({
    "event":["concert" , "conference" , "wedding"],
    "date":['2025-01-01' , '2025-03-15' , '2025-07-20']
})
```

```python
df["pandas_date"] =pd.to_datetime(df["date"])
```
